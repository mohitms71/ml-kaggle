```python
df_new2=pd.merge(df_rating,df_movie,on=['movieId'])
df_new2.head(20)
# print(df_new2[df_new2['movieId']==73881])
```

```python
movie_rating_count=df_new2.groupby(['title'])['rating'].count().reset_index().rename(columns={'rating':'totalrating'})
df_temp=pd.merge(df_new2,movie_rating_count,left_on=['title'],right_on=['title'],how='left')
df_temp.head(20)
```

```python
def create_user_movie_df():
    rating_counts = pd.DataFrame(df_new2["title"].value_counts())
    rating_counts.describe([0.05, 0.50, 0.75, 0.85, 0.90, 0.95, 0.99]).T
    limit = rating_counts.quantile([0.90]).T
    limit_90 = limit[0.9][0]
    rare_movies = rating_counts[rating_counts["title"] <= limit_90].index
    common_movies =df_new2[~df_new2["title"].isin(rare_movies)]
    user_movie_df = common_movies.pivot_table(index="userId", columns="title", values="rating")
    return user_movie_df
```

```python
table=create_user_movie_df().fillna(0)
# table=rating_popular.pivot_table(index="userId", columns="title", values="rating")
table.head()
# table.loc['Primer (2004)',:]
```
