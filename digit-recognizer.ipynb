```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
s=df['label']
print(s.shape)
y=pd.get_dummies(s)
print(s.head())
print(y.head())
#y=OneHotEncoder().fit(y)
x=df.drop(columns=['label'], axis=1).astype('float')/255.0
x_train,x_dev,y_train,y_dev=train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape,x_dev.shape,y_train.shape,y_dev.shape)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
x_train=tf.convert_to_tensor(x_train, dtype=tf.float32)
x_dev=tf.convert_to_tensor(x_dev, dtype=tf.float32)
y_train=tf.convert_to_tensor(y_train, dtype=tf.float32)
y_dev=tf.convert_to_tensor(y_dev, dtype=tf.float32)
#class mycallback(tf.keras.callbacks.Callback):
#    def on_epoch_end(self,epoch,logs={}):
#        if(logs.get('accuracy')>0.99):
#            print("\nReached accuracy 99%")
 #            self.model.stop_training=True
#callbacks=mycallback()
model=keras.Sequential([
    layers.Dense(128,activation=tf.nn.relu),
    layers.Dense(10,activation=tf.nn.softmax)
])
model.compile(optimizer=RMSprop(lr=0.001),loss=['mse'],metrics=['acc'])
model.fit(x_train,y_train,epochs=20)
dev_loss, dev_acc = model.evaluate(x_dev,  y_dev, verbose=2)

print(dev_loss,dev_acc)
```
